Импортируем необходиме библиотеки:

```python
import pandas as pd
import seaborn as sns

from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
```

Подключим набор данных:

```python
PATH = 'creditcard_2023_reduced.csv'
dataset = pd.read_csv(PATH)
```

Быстрая, визуальная проверка на коллинеарность:

```python
sns.heatmap(
    round(
        abs(dataset.corr()),
        1
    ),
    annot=True
)
```

Разделим выборку на тренировчную и оценочную часть, а также на параметры и эталоны:

```python
train_input, test_input, train_output, test_output = train_test_split(
    dataset.drop(['Class', 'id'], axis=1),
    dataset['Class'],
    test_size=0.2
)
```

Обучение, тестирование и проверка модели:

```python
model = GaussianNB()
model.fit(train_input, train_output)

predictions = model.predict(test_input)
accuracy = metrics.accuracy_score(predictions, test_output)

print(f"Точность модели на тестовом участке: {accuracy}")
```

Точность модели на тестовой части набора составила 0.99.
