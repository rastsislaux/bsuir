## –ó–∞–¥–∞–Ω–∏–µ 2: –ü—Ä–æ–≤–µ—Ä–∫–∞ –∫–æ–Ω—Å–∏—Å—Ç–µ–Ω—Ç–Ω–æ—Å—Ç–∏ –¥–∞–Ω–Ω—ã—Ö.

–ù–∞–±–æ—Ä –¥–∞–Ω–Ω—ã—Ö –¥–ª—è –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è: "Climate Change: Earth Surface Temperature Data" –Ω–∞ Kaggle. 

–°—Å—ã–ª–∫–∞: https://www.kaggle.com/berkeleyearth/climate-change-earth-surface-temperature-data



```python
files = [
    'GlobalLandTemperaturesByCity',
    'GlobalLandTemperaturesByCountry',
    'GlobalLandTemperaturesByMajorCity',
    'GlobalTemperatures',
    'GlobalLandTemperaturesByState'
]

datasets = {}

for file in files:
    datasets[file] = pd.read_csv(file + ".csv")

datasets['GlobalLandTemperaturesByCity']
```

```python
for file, df in datasets.items():
    df['dt'] = pd.to_datetime(df['dt'], errors='coerce')

    incorrect_dates = df[df['dt'].isnull()]
    if len(incorrect_dates) > 0:
        print(f'–í —Ñ–∞–π–ª–µ {file} –Ω–∞–π–¥–µ–Ω—ã –Ω–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–µ –¥–∞—Ç—ã:')
        print(incorrect_dates)
    else:
        print(f'–í—Å–µ –∑–∞–ø–∏—Å–∏ –≤ {file} –∏–º–µ—é—Ç –∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–π —Ñ–æ—Ä–º–∞—Ç.')
```

–†–∞—Å—Å–º–æ—Ç—Ä–∏—Ç–µ –Ω–∞–±–æ—Ä –¥–∞–Ω–Ω—ã—Ö –æ –∑–∞–±–æ–ª–µ–≤–∞–µ–º–æ—Å—Ç–∏ —Ä–∞–∫–æ–º, –ø—Ä–æ–∞–Ω–∞–ª–∏–∑–∏—Ä—É–π—Ç–µ, –ø—Ä–∏–≤–µ–¥–µ–Ω—ã –ª–∏ –¥–∞–Ω–Ω—ã–µ –ø–æ –≤—Å–µ–º —Å—Ç—Ä–∞–Ω–∞–º –∏–ª–∏ –Ω–µ–∫–æ—Ç–æ—Ä—ã–µ –∏–∑ –Ω–∏—Ö –æ—Ç—Å—É—Ç—Å—Ç–≤—É—é—Ç. 

–ù–∞–±–æ—Ä –¥–∞–Ω–Ω—ã—Ö –¥–ª—è –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è: "Global Cancer Statistics 2018 (GLOBOCAN 2018)" –Ω–∞ Kaggle.

–°—Å—ã–ª–∫–∞: https://www.kaggle.com/ashkhagan/global-cancer-statistics-2018-globocan-2018


```python
df = pd.read_csv("Countries.csv")
df
```

```python
unique_countries = df['Country Name'].unique()

dataset = df.copy()
dataset.dropna(inplace=True)
new_unique_countries = dataset['Country Name'].unique()

total_countries = len(unique_countries)

print(f"–í—Å–µ–≥–æ —Å—Ç—Ä–∞–Ω: {total_countries}")
print(f"–ü—Ä–∏–≤–µ–¥–µ–Ω–æ –¥–∞–Ω–Ω—ã—Ö –ø–æ —Å—Ç—Ä–∞–Ω–∞–º: {len(new_unique_countries)}")
print(f"–ü—Ä–∏–≤–µ–¥–µ–Ω—ã –¥–∞–Ω–Ω—ã–µ –ø–æ –≤—Å–µ–º —Å—Ç—Ä–∞–Ω–∞–º: {total_countries == len(new_unique_countries)}")

skipped_countries = []
for country in unique_countries:
    if country not in new_unique_countries:
        skipped_countries.append(country)


print("–°—Ç—Ä–∞–Ω—ã —Å –Ω–µ–ø–æ–ª–Ω—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏: " + ", ".join(skipped_countries))
```
